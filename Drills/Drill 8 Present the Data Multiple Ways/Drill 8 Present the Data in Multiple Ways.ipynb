```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#chunking for reading large csv files
chunks = pd.read_csv('agg_match_stats_0.csv',chunksize=1000000,sep=',')
chunk_tracker = []
for chunk in chunks:
    chunk_tracker.append(chunk)
```

```python
pubg = pd.DataFrame()
pubg = pd.concat(chunk_tracker)
```

```python
display(pubg)
```

```python
def remove_outliers(df,col):
    return df[np.abs(df[col]-df[col].mean())<=(3*df[col].std())]
```

```python
import calendar
pubg["date"] = pd.to_datetime(pubg["date"])
pubg["month"] = pubg["date"].dt.month
pubg["day"] = pubg["date"].dt.day
#pubg["month"] = pubg["month"].apply(lambda x: calendar.month_abbr[x])
pubg.head(5)
```

### II) Choose two continuous variables, and plot them three different ways (player_dmg & player_kills)

```python
# scatterplot (use player_survive_time instead of player_kills)
plyr_dmg_surv = pd.DataFrame()
plyr_dmg_surv["player_dmg"] = pubg["player_dmg"]
plyr_dmg_surv["month"] = pubg["month"]
plyr_dmg_surv["plyr_kills"] = pubg["player_kills"]
plyr_dmg_surv["team_id"] = pubg["team_id"]
plyr_dmg_surv["player_survive_time"] = pubg["player_survive_time"]
```

```python
a = plyr_dmg_surv["player_dmg"]
b = plyr_dmg_surv["player_survive_time"]

plt.scatter(a,b,alpha=0.5)
plt.xlabel("Player Damage")
plt.ylabel("Player Survive Time")
plt.title("PUBG Player Damage vs. Player Kills")
plt.show()
```

```python
# remove outliers
plyr_dmg_surv = remove_outliers(plyr_dmg_surv,"player_dmg")
plyr_dmg_surv = remove_outliers(plyr_dmg_surv,"player_survive_time")
# remove duplicate rows
plyr_dmg_surv = plyr_dmg_surv.drop_duplicates()
```

```python
# take random sample of population of data (1000) because dataset is large
plyr_dmg_surv_sample = plyr_dmg_surv.sample(1000,replace=True)
x_sample,y_sample = plyr_dmg_surv_sample["player_dmg"],plyr_dmg_surv_sample["player_survive_time"]
```
