### 1) What are the three longest trips on rainy days?

SELECT

    trips.trip_id,
  
    trips.duration,
  
    weather.Events
  
FROM

    trips
  
JOIN

    weather
  
ON

    trips.zip_code = weather.ZIP
  
WHERE

    Events = "Rain"
  
ORDER BY duration desc

LIMIT 3;

```python
trip_id     duration    Events
----------  ----------  ----------
1210487     83915       Rain
1210487     83915       Rain
1210487     83915       Rain
```

### 2) Which station is full most often?

SELECT

    station_id, docks_available, count(*) as dock_full_count

FROM

    status

WHERE

    docks_available = 0 
    
GROUP BY

    station_id

ORDER BY

    dock_full_count desc

LIMIT 1;

```python
station_id  docks_available  dock_full_count
----------  ---------------  ---------------
70          0                23450
```

### 3) Return a list of stations with a count of number of trips starting at that station but ordered by dock count.

    trips.start_station,

FROM

    trips

JOIN 

    stations

ON

    trips.start_station = stations.name

GROUP BY 1

### 4) (Challenge) What's the length of the longest trip for each day it rains anywhere?

SELECT

    max(trips.duration) max_duration, weather.Date, weather.Events

FROM

    trips
    
JOIN

    weather
    
ON

    trips.zip_code = weather.ZIP
    
WHERE

    weather.Events = "Rain"
    
GROUP BY weather.Date;
