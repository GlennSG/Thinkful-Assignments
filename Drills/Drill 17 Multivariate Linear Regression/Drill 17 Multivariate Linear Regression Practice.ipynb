```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
from sklearn import linear_model
%matplotlib inline
plt.style.use('seaborn')
sns.set_style('white')
```

```python
data = pd.read_csv('https://tf-curricula-prod.s3.amazonaws.com/data-science/Advertising.csv')
display(data)
```

```python
data.isnull().sum()
```

```python
regr = linear_model.LinearRegression()
Y = data["Sales"].values.reshape(-1,1)
X = data[["TV","Radio","Newspaper"]]
regr.fit(X,Y)

print("Coefficients: \n", regr.coef_)
print("\nIntercept: \n", regr.intercept_)
print("\nR-squared:")
print(regr.score(X,Y))
```

```python
# Extract predicted values
predicted = regr.predict(X).ravel()
actual = data["Sales"]

# Calculate error (residual)
residual = actual - predicted

plt.hist(residual)
plt.title("Residual counts")
plt.xlabel("Residual")
plt.ylabel("Count")
plt.show()
```

```python
plt.scatter(predicted,residual)
plt.xlabel("Predicted")
plt.ylabel("Residual")
plt.title('Residual vs Predicted')
plt.show()
```

```python
col_list = data.iloc[:,1:4].columns.values.tolist()
print(col_list)
```

```python
def normalize_data(df):
    x = (df-df.min())/(df.max()-df.min())
    return x

df_test = pd.DataFrame(columns = col_list)
for col in col_list:
    df_test[col] = normalize_data(data[col])
df_test.head(5)
```

```python
plt.rcParams['figure.figsize'] = (23,13)
fig,ax = plt.subplots(ncols = 3)
y = 0
for col in col_list:
    ax[y].boxplot(data[col],patch_artist=True)
    ax[y].set_title(col + " Distribution")
    y+=1
plt.show()
```

```python
# drop row with 0 value
df_test = df_test[(df_test[df_test_col] != 0).all(axis=1)]
```

```python
for col in df_test_col:
    print(0 in df_test[col].values)
```

```python
# transform data distributions to normal distribution
from scipy import stats
def transform_df(df_col):
    x,lam = stats.boxcox(df_col)
    return x

for col in df_test_col:
    df_test[col] = transform_df(df_test[col])
```

```python
fig,ax = plt.subplots(ncols = 3)
y = 0
for col in df_test_col:
    ax[y].hist(df_test[col])
    ax[y].set_title(col + " Distribution")
    y+=1
plt.show()
```

```python
# apply transformation to original dataframe
data = data[(data[df_test_col] != 0).all(axis=1)]

for col in col_list:
    data[col] = transform_df(data[col])
```
