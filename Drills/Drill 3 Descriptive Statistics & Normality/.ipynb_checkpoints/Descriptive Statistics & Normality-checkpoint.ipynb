```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

#### 1) Normal Distribution

```python
rand_norm_dist = np.random.normal(20,25,1500)
rand_norm_dist.sort()

plt.hist(rand_norm_dist,bins=20,color="c")
plt.axvline(rand_norm_dist.mean(),color="b",linestyle="solid",linewidth=2)
plt.axvline(rand_norm_dist.mean() + rand_norm_dist.std(),color="b",linestyle="dashed",linewidth=2)
plt.axvline(rand_norm_dist.mean() - rand_norm_dist.std(),color="b",linestyle="dashed",linewidth=2)
plt.show()
```

#### 2) Poisson Distribution

```python
rand_pois_dist = np.random.poisson(2,1500)
rand_pois_dist.sort()

plt.hist(rand_pois_dist,bins=20,color="c")
plt.axvline(rand_pois_dist.mean(),color="b",linestyle="solid",linewidth=2)
plt.axvline(rand_pois_dist.mean() + rand_pois_dist.std(),color="b",linestyle="dashed",linewidth=2)
plt.axvline(rand_pois_dist.mean() - rand_pois_dist.std(),color="b",linestyle="dashed",linewidth=2)
plt.show()
```

#### 3) Bernoulli Distribution

```python
rand_bern_dist = np.random.binomial(1,.5,1500)
rand_bern_dist.sort()

plt.hist(rand_bern_dist,bins=20,color="c")
plt.axvline(rand_bern_dist.mean(),color="b",linestyle="solid",linewidth=2)
plt.axvline(rand_bern_dist.mean() + rand_bern_dist.std(),linestyle="dashed",color="b",linewidth=2)
plt.axvline(rand_bern_dist.mean() - rand_bern_dist.std(),linestyle="dashed",color="b",linewidth=2)
plt.show()
```

#### 4) Binomial Distribution

```python
rand_binomial = np.random.binomial(50,0.5,1500)
rand_binomial.sort()

plt.hist(rand_binomial,bins=20,color="c")
plt.axvline(rand_binomial.mean(),color="b",linestyle="solid",linewidth=2)
plt.axvline(rand_binomial.mean() + rand_binomial.std(),linestyle="dashed",color="b",linewidth=2)
plt.axvline(rand_binomial.mean() - rand_binomial.std(),linestyle="dashed",color="b",linewidth=2)
plt.show()
```

#### 5) Gamma Distribution

```python
rand_gamma = np.random.gamma(5,1,1500)
rand_gamma.sort()

plt.hist(rand_gamma,bins=20,color="c")
plt.axvline(rand_gamma.mean(),color="b",linestyle="solid",linewidth=2)
plt.axvline(rand_gamma.mean() + rand_gamma.std(),color="b",linestyle="dashed",linewidth=2)
plt.axvline(rand_gamma.mean() - rand_gamma.std(),color="b",linestyle="dashed",linewidth=2)
plt.show()
```

#### 6) Conditional Distribution

```python
budget = pd.DataFrame()
budget["spending"] = np.random.normal(100,20,2000)
budget["cost"] = np.random.randint(20,1000,2000)

condition = budget[budget["cost"] > 200]["spending"]
plt.hist(condition,bins=20,color="c")
plt.axvline(condition.mean(),color="b",linestyle="solid",linewidth=2)
plt.axvline(condition.mean() + condition.std(),color="b",linestyle="dashed",linewidth=2)
plt.axvline(condition.mean() - condition.std(),color="b",linestyle="dashed",linewidth=2)
plt.show()
```

#### Extra Assignment:

```python
rand_norm1 = np.random.normal(5,0.5,1000)
rand_norm2 = np.random.normal(10,1,1000)
rand_norm3 = rand_norm1 + rand_norm2
rand_norm3.sort()
plt.hist(rand_norm3,bins=20,color="c")
plt.axvline(rand_norm3.mean(),color="b",linestyle="solid",linewidth=2)
plt.axvline(rand_norm3.mean() + rand_norm3.std(),color="b",linestyle="dashed",linewidth=2)
plt.axvline(rand_norm3.mean() - rand_norm3.std(),color="b",linestyle="dashed",linewidth=2)
plt.show()
```
